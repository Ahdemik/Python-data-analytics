#Merge,Join, and Concatenate

```python
import pandas as pd
```

```python
df1= pd.read_excel(r"C:\Users\HP Pavilion\Desktop\Data analysis\Practices\personal project\Dataset for Portfolio\Table 1 for merging practise.xlsx")
df1
```

```python
df2= pd.read_excel(r"C:\Users\HP Pavilion\Desktop\Data analysis\Practices\personal project\Dataset for Portfolio\Table 2 for merging practise.xlsx")
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner', on = 'EmployeeID')
```

```python
df1.merge(df2, how = 'inner', on = ['EmployeeID','FirstName', 'LastName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
#df1.merge(df2, how = 'cross')
```

```python
df3= df1.set_index('EmployeeID').join(df2.set_index('EmployeeID'), lsuffix = '_left',rsuffix='_Right', how = 'inner')
df3
```

```python
df3= df1.set_index(['EmployeeID','FirstName','LastName']).join(df2.set_index(['EmployeeID','FirstName','LastName']), lsuffix = '_left',rsuffix='_Right', how = 'inner')
df3
```

```python
df3= df1.set_index(['EmployeeID','FirstName','LastName']).join(df2.set_index(['EmployeeID','FirstName','LastName']), lsuffix = '_left',rsuffix='_Right', how = 'outer')
df3
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join='inner')
```

```python
pd.concat([df1,df2], join='inner', axis=1)
```
