```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Nigerian_states_by_GDP'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('th')[0:4]
```

```python
heading = soup.find_all('th')[0:4]
```

```python
print(heading)
```

```python
heading= [heading_.text.strip() for heading_ in heading]
```

```python
print(heading)
```

```python
import pandas as pd

df = pd.DataFrame(columns = heading)

df
```

```python
soup.find_all('tr')
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data[1:37]:
    row_data = row.find_all('td')[0:4]
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data





```
